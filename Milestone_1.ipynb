```python
# -----------------------------
# Step 1: Install required library
# -----------------------------
!pip install --quiet google-api-python-client
print("Library installed successfully!")

# -----------------------------
# Step 2: Import libraries
# -----------------------------
from googleapiclient.discovery import build
import pandas as pd

# -----------------------------
# Step 3: Set API key and Channel ID
# -----------------------------
api_key = "AIzaSyD2mTSj7if6dJpoUwaXfHh5O9dmw3IE3fo"          # Replace with your YouTube Data API key
channel_id = "UC4rlAVgAK0SGk-yTfe48Qpw"  # Replace with the channel ID
max_results = 10                 # Number of latest videos to fetch

# -----------------------------
# Step 4: Build YouTube service
# -----------------------------
youtube = build('youtube', 'v3', developerKey=api_key)

# -----------------------------
# Step 5: Fetch latest videos from the channel
# -----------------------------
search_request = youtube.search().list(
    part='snippet',
    channelId=channel_id,
    maxResults=max_results,
    order='date'
)
search_response = search_request.execute()

# -----------------------------
# Step 6: Extract video IDs
# -----------------------------
video_ids = [item['id']['videoId'] for item in search_response['items'] if item['id']['kind'] == 'youtube#video']

# -----------------------------
# Step 7: Fetch video details
# -----------------------------
video_request = youtube.videos().list(
    part='snippet,contentDetails,statistics',
    id=','.join(video_ids)
)
video_response = video_request.execute()

# -----------------------------
# Step 8: Prepare data for CSV
# -----------------------------
data = []
for video in video_response['items']:
    data.append({
        "Video ID": video['id'],
        "Title": video['snippet']['title'],
        "Description": video['snippet']['description'],
        "Published At": video['snippet']['publishedAt'],
        "Duration": video['contentDetails']['duration'],
        "Views": video['statistics'].get('viewCount', 'N/A')
    })

df = pd.DataFrame(data)

# -----------------------------
# Step 9: Save data to CSV
# -----------------------------
csv_file = "youtube_videos.csv"
df.to_csv(csv_file, index=False)
print(f"Video details saved to {csv_file}")

# -----------------------------
# Step 10: Display the data
# -----------------------------
df
```

```python
# Step 1: Install the Google API client
!pip install --quiet google-api-python-client pandas

# Step 2: Import required libraries
from googleapiclient.discovery import build
import pandas as pd

# Step 3: Set up your YouTube API credentials
API_KEY = 'AIzaSyD2mTSj7if6dJpoUwaXfHh5O9dmw3IE3fo'  # <-- Replace with your actual API key
CHANNEL_ID = 'UC4rlAVgAK0SGk-yTfe48Qpw'  # <-- Replace with the target channel ID

youtube = build('youtube', 'v3', developerKey=API_KEY)

# Step 4: Function to fetch all videos using pagination
def get_all_videos(channel_id):
    videos = []
    next_page_token = None

    while True:
        # Fetch playlist ID for uploaded videos
        res = youtube.channels().list(
            part='contentDetails',
            id=channel_id
        ).execute()

        uploads_playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']

        # Fetch videos from playlist
        playlist_res = youtube.playlistItems().list(
            part='snippet',
            playlistId=uploads_playlist_id,
            maxResults=50,  # maximum allowed by API
            pageToken=next_page_token
        ).execute()

        for item in playlist_res['items']:
            video_id = item['snippet']['resourceId']['videoId']
            title = item['snippet']['title']
            published_at = item['snippet']['publishedAt']
            videos.append({
                'video_id': video_id,
                'title': title,
                'published_at': published_at
            })

        next_page_token = playlist_res.get('nextPageToken')
        if not next_page_token:
            break

    return videos

# Step 5: Run function and save to CSV
video_list = get_all_videos(CHANNEL_ID)
df = pd.DataFrame(video_list)
df.to_csv('youtube_channel_videos.csv', index=False)

print(f"✅ {len(video_list)} videos fetched and saved to 'youtube_channel_videos.csv'.")
```
